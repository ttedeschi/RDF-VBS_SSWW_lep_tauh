```python
import ROOT
import os

QCD_1_file = ROOT.TFile.Open("../RDataFrame_porting/out_QCDHT_100to200_ptw07.root","READ")
QCD_2_file = ROOT.TFile.Open("../RDataFrame_porting/out_QCDHT_500to700_ptw07.root","READ")
QCD_3_file = ROOT.TFile.Open("../RDataFrame_porting/out_QCDHT_1000to1500_ptw07.root","READ")
VBS_file = ROOT.TFile.Open("../RDataFrame_porting/out_VBS_ptw07.root","READ")

variables = [
"leadingJet_pt",
"leadingJet_nContistuents",
"leadingJet_constituentEtaPhiSpread",    
"leadingJet_nHFHadrons",
"leadingJet_nHFemParticles",
"leadingJet_nElectrons",
"leadingJet_nMuons",
"leadingJet_nPhotons",
"leadingJet_nNeutralHadrons",
"leadingJet_nChargedHadrons",
"subleadingJet_nContistuents",
"subleadingJet_constituentEtaPhiSpread",
"subleadingJet_nHFHadrons",
"subleadingJet_nHFemParticles",
"subleadingJet_nElectrons",
"subleadingJet_nMuons",
"subleadingJet_nPhotons",
"subleadingJet_nNeutralHadrons",
"subleadingJet_nChargedHadrons"]

canvases = []
legends = []

for i, v in enumerate(variables):
    canvases.append(ROOT.TCanvas())
    canvases[i].SetLogy()
    variable = variables[i]
    #h_qcd_1 = QCD_1_file.Get("{}".format(variable))
    h_qcd_2 = QCD_2_file.Get("{}".format(variable))
    h_qcd_3 = QCD_3_file.Get("{}".format(variable))
    h_vbs = VBS_file.Get("{}".format(variable))
    #h_qcd_1.Scale( 1./h_qcd_1.Integral(),"WIDTH")
    h_qcd_2.Scale( 1./h_qcd_2.Integral(),"WIDTH")
    h_qcd_3.Scale( 1./h_qcd_3.Integral(),"WIDTH")
    h_vbs.Scale( 1./h_vbs.Integral(),"WIDTH")
    #maximum = max(h_qcd_1.GetMaximum(), h_qcd_2.GetMaximum(), h_qcd_3.GetMaximum(), h_vbs.GetMaximum())
    maximum = max(h_qcd_2.GetMaximum(), h_qcd_3.GetMaximum(), h_vbs.GetMaximum())
    #h_qcd_1.SetLineColor(3)
    #h_qcd_1.GetYaxis().SetRangeUser(0.0000001,maximum)
    #h_qcd_1.SetStats(False)
    h_qcd_2.SetLineColor(2)
    h_qcd_2.GetYaxis().SetRangeUser(0.0000001,maximum)
    h_qcd_2.SetStats(False)
    h_qcd_3.SetLineColor(11)
    h_qcd_3.GetYaxis().SetRangeUser(0.0000001,maximum)
    h_qcd_3.SetStats(False)
    h_vbs.SetStats(False)
    h_vbs.GetYaxis().SetRangeUser(0.0000001,maximum)
    h1 = h_vbs.DrawCopy('HIST')
    #h2_1 = h_qcd_1.DrawCopy('HIST SAME')
    h2_2 = h_qcd_2.DrawCopy('HIST SAME')
    h2_3 = h_qcd_3.DrawCopy('HIST SAME')
    legends.append(ROOT.TLegend(.73,.32,.97,.53))
    legends[i].SetBorderSize(0)
    legends[i].SetFillColor(0)
    legends[i].SetFillStyle(0)
    legends[i].SetTextFont(42)
    legends[i].SetTextSize(0.035)
    legends[i].AddEntry(h1, "VBS", "L")
    #legends[i].AddEntry(h2_1, "QCD_HT100to200", "L")
    legends[i].AddEntry(h2_2, "QCD_HT500to700", "L")
    legends[i].AddEntry(h2_3, "QCD_HT1000to1500", "L")
    legends[i].Draw()
    canvases[i].Draw()
    #c.Show()
```

```python
df = ROOT.RDataFrame("Events", "VBS_SSWW_LL/AA101ED1-3427-E44D-9D6E-82569A5A589C.root ")
df_atleast2Jets = df.Filter("nJet>2", "At least two jets")
df_GoodJets = df_atleast2Jets.Define("GoodJets_idx", "GoodJets(Jet_jetId, Jet_eta, Jet_pt, Jet_puId)")
df_atleast2GoodJets = df_GoodJets.Filter("atleast2GoodJets(GoodJets_idx)", "At least two good jets")
df_VBSjets = df_atleast2GoodJets.Define("VBSJet_idx", "SelectVBSJets_invmass(Jet_pt, Jet_eta, Jet_phi, Jet_puId, GoodJets_idx)")
df_2VBSjets = df_VBSjets.Filter("VBSJet_idx[0] != VBSJet_idx[1]", "2 VBS jets")
df_GenJets = df_2VBSjets.Define("GenJet_idx", "SelectVBSQGenJet(GenJet_partonFlavour)")
df_atleast2GenJets = df_GenJets.Filter("atleast2Jets(GenJet_idx)","At least 2 GenJets")
df_defineGenMatching = df_atleast2GenJets.Define("leadjet_IsGenMatched", "IsGenMatched(GenJet_idx[0], Jet_genJetIdx[VBSJet_idx[0]])")\
                                         .Define("subleadjet_IsGenMatched", "IsGenMatched(GenJet_idx[1], Jet_genJetIdx[VBSJet_idx[1]])")\
                                         #.Filter("HLT_IsoMu27 || HLT_Mu50 || HLT_Ele35_WPTight_Gsf || HLT_Ele32_WPTight_Gsf_L1DoubleEG || HLT_Photon200", "Leptonic trigger")

h = {}

h["leadjet_IsGenMatched"] = df_defineGenMatching.Histo1D(("leadjet_IsGenMatched", "leadjet_IsGenMatched", 2, 0, 2),"leadjet_IsGenMatched")
h["subleadjet_IsGenMatched"] = df_defineGenMatching.Histo1D(("subleadjet_IsGenMatched", "subleadjet_IsGenMatched", 2, 0, 2),"subleadjet_IsGenMatched")
```
